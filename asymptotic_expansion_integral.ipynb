# Asymptotic expansion of integrals

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams.update({'font.size': 18})
```

## laplace method

We consider a definite integral of the form:
$$
I(\lambda) = \int^b_a f(t) e^{\lambda g(t)}\,dt, \quad \lambda \gg 1.
$$

### Example

Here we choose 
$$
a = -1, \quad b = 1, \quad f(t) = \cos(t), \quad g(t) = \sin(\sin(3t)).
$$

Assuming that the maximum of $g$ occurs at $x=c$, we then define
$$
F = \frac{f(t)e^{\lambda g(t)}}{e^{\lambda g(c)}} = f(t)e^{\lambda (g(t)-g(c))}
$$
and visualize the behavior when $\lambda\gg 1$.

```python
x = np.linspace(-1,1, 1000)
f = np.cos(x)
g = np.sin(np.sin(3.0*x))
g = g - np.max(g)
```

```python
plt.plot(x, f, label='f')
plt.plot(x, g, label='g')
plt.xlabel('x')
plt.legend(loc = 4)
plt.show()
```

```python
plt.figure(figsize = (25,10))

plt.subplot(2, 3, 1)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')

plt.subplot(2, 3, 2)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')

plt.subplot(2, 3, 3)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')


plt.subplot(2, 3, 4)
lambd=1
F = f*np.exp(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=1')

plt.subplot(2, 3, 5)
lambd=10**3
F = f*np.exp(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=10^3')

plt.subplot(2, 3, 6)
lambd=10**6
F = f*np.exp(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=10^6')

plt.show()
```

### Observation

It is clear that the integrand, as $\lambda \to \infty$, has a peak located exactly at the location where $g(x)$ has a maximum, say, $x=c$. So the significant contribution to the integral will come only from points $x$ in a neighbourhood of $c$, which can then be estimated.

### Derivation

Assuming that $g$ has a local max. at $t=c$, so the Taylor expansion of $g(t)$ at $t=c$ is
$$
g(t) \approx g(c) + \frac{g''(c) t^2}{2},
$$
where we have used the fact that $g'(c)=0$, and note that $g''(c)<0$.

For $\lambda \gg 1$, 
$$
I(\lambda) = \int^b_a f(t) e^{\lambda g(t)}\,dt \approx \int^{c+\epsilon}_{c-\epsilon} f(t) e^{\lambda g(t)}\,dt \approx f(c)e^{\lambda g(c)} \int^{c+\epsilon}_{c-\epsilon} e^{\lambda g''(c)/2}\,dt \approx f(c)e^{\lambda g(c)} \int^{\infty}_{-\infty} e^{\lambda g''(c)t^2/2}\,dt = f(c)e^{\lambda g(c)}\sqrt{\frac{2\pi}{\lambda |g''(c)|}}.
$$

## Watson's lemma

We consider a definite integral of the form:
$$
I(\lambda) = \int^b_0 t^{\alpha}h(t) e^{-\lambda t}\,dt, \quad \lambda \gg 1.
$$

### Example

Here we choose 
$$
b=1, \quad \alpha = 0, \quad h(t) = \cos(t),
$$
and define 
$$
F = t^{\alpha}h(t) e^{-\lambda t}
$$
to visualize the behavior when $\lambda\gg 1$. 

```python
x = np.linspace(0,1, 1000)
h = np.cos(x)
```

```python
lambd = 1.0;


plt.figure(figsize = (25,4))

plt.subplot(1, 3, 1)
lambd = 1.0
F = h*np.exp(-lambd*x);
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=1')

plt.subplot(1, 3, 2)
lambd = 10.0
F = h*np.exp(-lambd*x);
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=10')

plt.subplot(1, 3, 3)
lambd=100.0
F = h*np.exp(-lambd*x);
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=100')

plt.show()
```

### Observation

It is clear that, as $\lambda \to \infty$, the significant contribution to the integral will come only from $x\approx 0$, which can then be estimated.

## Method of stationary phase

We consider a definite integral of the form:
$$
I(\lambda) = \int^b_a f(t) e^{i\lambda g(t)}\,dt, \quad \lambda \gg 1.
$$

Notice that 
$$
I(\lambda) = \int^b_a f(t) e^{i\lambda g(t)}\,dt = \int^b_a f(t)\left(\cos(\lambda g(t))+i\sin(\lambda g(t))\right)\,dt
$$

### Example

Here we choose 
$$
a = -1, \quad b = 1, \quad f(t) = \cos(t), \quad g(t) = \sin(\sin(3t)).
$$
and define 
$$
F = f(t) \cos(\lambda g(t))
$$
to visualize the behavior when $\lambda\gg 1$. 

```python
x = np.linspace(-1,1, 10000)
f = np.cos(x)
g = np.sin(np.sin(3.0*x))
```

```python
plt.figure(figsize = (25,10))

plt.subplot(2, 3, 1)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')

plt.subplot(2, 3, 2)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')

plt.subplot(2, 3, 3)
plt.plot(x, g)
plt.xlabel('x')
plt.ylabel('g')


plt.subplot(2, 3, 4)
lambd=10
F = f*np.cos(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=10')

plt.subplot(2, 3, 5)
lambd=10**2
F = f*np.cos(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=100')

plt.subplot(2, 3, 6)
lambd=10**3
F = f*np.cos(lambd*g)
plt.plot(x, F)
plt.xlabel('x')
plt.ylabel('F: lambda=1000')

plt.show()
```

### Observation

As $\lambda \to \infty$, the integrand exhibits strong oscillations except at points where $\frac{d}{dt} g=0$.
