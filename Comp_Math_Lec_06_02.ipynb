### Errors of polynomial interpolation

```julia
n=6;
xi1 = [-1; 2*rand(n-2)-1; 1]
```

```julia
x = linspace(-1,1,1000);
p1 = x-xi1[1];
for ii=1:n-1
    p1 = p1.*(x-xi1[ii+1]);
end
println(maximum(abs.(p1), 1))
using PyPlot
plot(x, abs.(p1), xi1, zeros(n,1), "or");
```

```julia
xi2 = zeros(6,1);
for ii=1:n
    xi2[ii] = xi1[ii];
end
xi2[3] = -0.5;
p2 = x-xi2[1];
for ii=1:n-1
    p2 = p2.*(x-xi2[ii+1]);
end
println(maximum(abs.(p2), 1))
using PyPlot
plot(x, abs.(p2), xi2, zeros(n,1), "or", x, abs.(p1));
```

```julia
xi_cheb = cos.(((1:n)-0.5)*pi/n)
```

```julia
x2 = linspace(-1,1,1000);
p_cheb = x2-xi_cheb[1];
for ii=1:n-1
    p_cheb = p_cheb.*(x2-xi_cheb[ii+1]);
end
println(maximum(abs.(p_cheb), 1))
using PyPlot
plot(x2, (p_cheb), xi_cheb, zeros(n,1), "or")
```
