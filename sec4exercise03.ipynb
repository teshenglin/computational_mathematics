```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.interpolate import barycentric_interpolate

pi=math.pi
def f(x):
  return(abs(x))
def Chebyshevpoints(i,n):
  t=math.cos(i*pi/n)
  return(-t)

fig=plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
#make data list
for k in range(5):
  n=(k+1)*10
  x=[]
  y=[]
  for i in range(n+1):
    t=Chebyshevpoints(i,n)
    x.append(t)
    y.append(f(x[i]))

#data for check poly
  xnew = np.linspace(-1.0, 1.0, 1000)
  fynew = f(xnew)

#Interpolation polynomial
  ynew = barycentric_interpolate(x, y, xnew)
  plt.plot(xnew, ynew)
  plt.plot(xnew, fynew, 'r')

plt.title('Interpolation')
plt.xlabel('x')
plt.legend(['n=10','n=20','n=30','n=40','n=50'], loc ="upper left")


plt.subplot(1, 2, 2)
#make data list
for k in range(5):
  n=(k+1)*10
  x=[]
  y=[]
  for i in range(n+1):
    t=Chebyshevpoints(i,n)
    x.append(t)
    y.append(f(x[i]))

#data for check poly
  xnew = np.linspace(-1.0, 1.0, 1000)
  fynew = f(xnew)

#Interpolation polynomial
  ynew = barycentric_interpolate(x, y, xnew)
  plt.plot(xnew, np.abs(ynew-fynew))

plt.title('Abs. error')
plt.xlabel('x')
plt.legend(['n=10','n=20','n=30','n=40','n=50'], loc ="upper left")

plt.show()
```
